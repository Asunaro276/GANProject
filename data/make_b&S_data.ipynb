```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from itertools import combinations_with_replacement
```

```python
N = 2
zeros = np.zeros((N, N))
```

```python
bars = []
for ind in combinations_with_replacement(range(N), N-1):
    bar = zeros.copy()
    for i in ind:
        bar[:, i] = 1
    bars.append(bar)
bars = np.unique(np.array(bars), axis=0)
bars
```

```python
strips = []
for ind in combinations_with_replacement(range(N), N-1):
    strip = zeros.copy()
    for i in ind:
        strip[i, :] = 1
    strips.append(strip)
strips = np.unique(np.array(strips), axis=0)
strips
```

```python
bars = bars.reshape((-1, N*N))
strips = strips.reshape((-1, N*N))
```

```python
dir_path = f"./img_bars_and_strips_{N}/img_b/"
os.makedirs(dir_path, exist_ok=True)
for i in range(2**N-2):
    img_path = os.path.join(dir_path, f"img_b_{i}.png")
    img = bars[np.newaxis, i]
    cv2.imwrite(img_path, img)
    plt.imshow(img)
    plt.show()

dir_path = f"./img_bars_and_strips_{N}/img_s/"
os.makedirs(dir_path, exist_ok=True)
for i in range(2**N-2):
    img_path = os.path.join(dir_path, f"img_s_{i}.png")
    img = strips[np.newaxis, i]
    cv2.imwrite(img_path, img)
    plt.imshow(img)
    plt.show()
```
