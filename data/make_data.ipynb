```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
from itertools import combinations_with_replacement
```

```python
array_list = []
for array in combinations_with_replacement([0, 1, 0, 1, 0], 8):
    array_list.append(array)
array_list = np.array(array_list, dtype=np.int8)
array_list = np.unique(array_list, axis=0)
np.random.shuffle(array_list)
array_list
```

```python
array_list_0 = []
for array in combinations_with_replacement([0, 1, 0], 8):
    array_list_0.append(array)
array_list_0 = np.array(array_list_0, dtype=np.int8)
array_list_0 = np.unique(array_list_0, axis=0)
array_list_0
```

```python
num_data = 28
search_array = lambda array, array_list: np.any(np.all(array == array_list, axis=1))
np.random.shuffle(array_list_0)
array_list_0 = np.array(array_list_0)
array_list_1 = []
for array in array_list:
    if not search_array(array, array_list_0):
        array_list_1.append(array)
    if len(array_list_1) == num_data:
        break
array_list_0 = array_list_0[:num_data]
array_list_1 = np.array(array_list_1)
```

```python
dir_path = "./img_qSGAN/img_0/"
for i in range(num_data):
    img_path = os.path.join(dir_path, f"img_0_{i}.png")
    img = array_list_0[np.newaxis, i]
    plt.imshow(img)
    plt.savefig(img_path)
    plt.show()

dir_path = "./img_qSGAN/img_1/"
for i in range(num_data):
    img_path = os.path.join(dir_path, f"img_1_{i}.png")
    img = array_list_1[np.newaxis, i]
    plt.imshow(img)
    plt.savefig(img_path)
    plt.show()
```
